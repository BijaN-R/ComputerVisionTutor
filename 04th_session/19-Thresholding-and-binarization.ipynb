```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

#image = cv2.imread('../03rd_session/images/jolge.jpg')
image = cv2.imread('./images/gradient_gray.PNG')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

thresh = 127

new_image = np.copy(image)
new_image[new_image < thresh] = 0
new_image[new_image >= thresh] = 255

plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title('Original');
plt.subplot(122);plt.imshow(new_image, cmap='gray');plt.title('Output');
```

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

image = cv2.imread('./images/gradient_gray.PNG', 0)
#image = cv2.imread('../03rd_session/images/jolge.jpg', 0)

ret, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret, binary_inv_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
ret, trunc_image = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
ret, tozero_image = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
ret, tozero_image_inv = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
print(ret)
plt.figure(figsize=[9,8])
plt.subplot(231);plt.imshow(image, cmap='gray');plt.title('Original');
plt.subplot(232);plt.imshow(binary_image, cmap='gray');plt.title('1-Binary');
plt.subplot(233);plt.imshow(binary_inv_image, cmap='gray');plt.title('2-Binary invert');
plt.subplot(234);plt.imshow(trunc_image, cmap='gray');plt.title('3-Truncate');
plt.subplot(235);plt.imshow(tozero_image, cmap='gray');plt.title('4-Tozero');
plt.subplot(236);plt.imshow(tozero_image_inv, cmap='gray');plt.title('5-Tozero invert');
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('./images/pastry-dish.jpg')
b, g, r = cv2.split(image)
```

```python
plt.figure(figsize=[20, 5])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title('Red Channel');
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title('Green Channel');
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title('Blue Channel');

plt.subplot(144); plt.imshow(image[..., ::-1]); plt.title('All Channels');
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('./images/pastry-dish.jpg')
b, g, r = cv2.split(image)

zeroes =np.zeros_like(r, 'uint8')

modified_blue = b.astype('int16') + 70
modified_blue = np.minimum(modified_blue, 255)
modified_blue = modified_blue.astype('uint8')

merged = cv2.merge([b, g, r])
modified_merged = cv2.merge([modified_blue, g, r])

plt.figure(figsize=[20, 10])
plt.subplot(241); plt.imshow(r, cmap= 'gray'); plt.title('Red');
plt.subplot(242); plt.imshow(g, cmap= 'gray'); plt.title('Green');
plt.subplot(243); plt.imshow(b, cmap= 'gray'); plt.title('Blue');
plt.subplot(244); plt.imshow(merged[..., ::-1]); plt.title('Oringinal Image');
plt.subplot(247); plt.imshow(modified_blue, cmap= 'gray'); plt.title('Modified Red');
plt.subplot(248); plt.imshow(modified_merged[..., ::-1]); plt.title('Modified Output');
```

```python
#h: 0 - 180, s: 0 - 255, v: 0 - 255

image = cv2.imread('./images/jolge.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

cv2.imshow('Original Image', image)
cv2.imshow('HSV Image', hsv_image)
cv2.imshow('Hue Image', hsv_image[..., 0])
cv2.imshow('Saturation Image', hsv_image[..., 1])
cv2.imshow('Value Image', hsv_image[..., 2])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image = cv2.imread('./images/jolge.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
saturated_image = cv2.merge([hsv_image[..., 0], cv2.add(hsv_image[..., 1], 50), hsv_image[..., 2]])
saturated_image = cv2.cvtColor(saturated_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[10, 5])
plt.subplot(121); plt.imshow(saturated_image[..., ::-1]); plt.title('Sturated');
plt.subplot(122); plt.imshow(image[..., ::-1]); plt.title('Original');
```

```python
image = cv2.imread('./images/jolge.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
sturated_mask = hsv_image[..., 2] < 230
inverted_mask = hsv_image[..., 2] >= 230
hsv_image[..., 1][sturated_mask] += 80
hsv_image[..., 1][inverted_mask] = 15
saturated_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[10, 5])
plt.subplot(121); plt.imshow(saturated_image[..., ::-1]); plt.title('Sturated');
plt.subplot(122); plt.imshow(image[..., ::-1]); plt.title('Original');
```

```python
image = cv2.imread('./images/jolge.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
brighter_image = cv2.merge([hsv_image[..., 0], hsv_image[..., 1], cv2.add(hsv_image[..., 2], 35)])
brighter_image = cv2.cvtColor(brighter_image, cv2.COLOR_HSV2BGR)

plt.figure(figsize=[10, 5])
plt.subplot(121); plt.imshow(brighter_image[..., ::-1]); plt.title('Brighter');
plt.subplot(122); plt.imshow(image[..., ::-1]); plt.title('Original');
```

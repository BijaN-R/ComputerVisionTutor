```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./media/portrait.jpg')
cropped = image[70:270, 320:480]
cv2.imwrite('./media/cropped.jpg', cropped)

plt.figure(figsize=[10,5])
plt.subplot(121); plt.imshow(image[...,::-1]); plt.title('Original');
plt.subplot(122); plt.imshow(cropped[..., ::-1]); plt.title('Cropped');
```

```python
import cv2
import math 
import numpy as np 
import matplotlib.pyplot as plt

color = (80, 255, 0)

def triangle(image, centerPoint, sideLen, color):
    heightBigHalf = round((sideLen/2) / math.cos(30*math.pi/180))
    heightShortHalf = round((sideLen/2) * math.tan(30*math.pi/180))
    a = [centerPoint[0], centerPoint[1] - heightBigHalf]
    b = [centerPoint[0] + round(sideLen/2), centerPoint[1] + heightShortHalf]
    c = [centerPoint[0] - round(sideLen/2), centerPoint[1] + heightShortHalf]
    points = np.array([a ,b ,c])
    points = points.reshape(-1, 1, 2)
    cv2.polylines(image, [points], True, color, 2)

def draw_triangle(event, x, y, flag, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        global color
        triangle(img, (x, y), 50, color)

img = np.zeros((512, 512, 3), np.uint8)

cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_triangle)

while True:
    cv2.imshow('image', img)
    k = cv2.waitKey(1) & 0xFF
    
    if k == ord('b'):
        color = (255, 50, 0)
    elif k == ord('g'):
        color = (80, 255, 0)
    elif k == ord('r'):
        color = (0, 0, 255)
    elif k == ord('o'):
        color = (0, 100, 255)
    elif k == ord('m'):
        color = (255, 0, 255)
    elif k == 27:
        break

cv2.destroyAllWindows()
plt.imshow(img[...,::-1])
```

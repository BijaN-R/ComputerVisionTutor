```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('./image/adad.jpg', 0)
blured = cv2.GaussianBlur(image, (15,15), 0) 

block_size = 11
c = 2 

output = cv2.adaptiveThreshold(blured, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, c)

plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title('Original')
plt.subplot(122);plt.imshow(output, cmap='gray');plt.title('Binary Image')

```

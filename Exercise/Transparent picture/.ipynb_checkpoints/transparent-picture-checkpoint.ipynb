```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./media/portrait.jpg')

height, width, _ = image.shape

transparent_img = np.zeros((height, width, 4), np.uint8)
transparent_img[..., 0:3] = image
transparent_img[..., 3] = 80
transparent_img[70:270, 320:480, 3] = 255

cv2.imwrite('./media/transparent_picture.png', transparent_img)

transparent_img = cv2.cvtColor(transparent_img, cv2.COLOR_BGRA2RGBA)

plt.figure(figsize=[15,5])
plt.subplot(131); plt.imshow(image[...,::-1]); plt.title('Original');
plt.subplot(132); plt.imshow(transparent_img[..., 3], 'gray'); plt.title('Mask');
plt.subplot(133); plt.imshow(transparent_img); plt.title('Transparent image');
```

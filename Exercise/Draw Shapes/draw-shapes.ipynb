```python
import cv2
import numpy as np
import matplotlib.pyplot as pt
```

```python
image = np.zeros((512, 512, 3), np.uint8)
image += 255
cv2.rectangle(image, (69, 229), (441, 440), (0, 0, 0), 4)
cv2.rectangle(image, (210, 300), (290, 440), (0, 0, 0), 4)
cv2.rectangle(image, (330, 300), (390, 360), (0, 0, 0), 4)
cv2.rectangle(image, (110, 300), (170, 360), (0, 0, 0), 4)

pts = np.array([[40, 229], [255, 100], [470, 229]])
pts = pts.reshape(-1, 1, 2)
cv2.polylines(image, [pts], True, (0, 0, 0), 4)

cv2.imwrite('./image/house.png', image)

pt.imshow(image[..., ::-1])
```

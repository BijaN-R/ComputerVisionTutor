```python
import cv2
import numpy as np
import matplotlib.pyplot as pt
```

```python
image = np.zeros((512, 512, 3), np.uint8)
image_bw = np.zeros((512, 512), np.uint8)

cv2.line(image, (0, 0), (511, 511), (250, 100, 0), 5)
cv2.line(image, (0, 511), (511, 0), (250, 100, 0), 5)
cv2.rectangle(image, (49, 49), (461, 461), (70, 130, 255), -5)
cv2.rectangle(image, (129, 129), (381, 381), (70, 255, 200), -5)
cv2.circle(image, (255, 255), 100, (100, 50, 255), -1)
cv2.ellipse(image, (255, 255), (250, 150), 45, 0, 360, (50, 150, 0), 5)
cv2.ellipse(image, (255, 255), (250, 150), -45, 0, 360, (50, 150, 0), 5)

cv2.imwrite('./images/Drawing-01.png', image)

pt.imshow(image[..., ::-1])
pt.figure()
pt.imshow(image_bw, cmap='gray')
```

```python
image = np.zeros((512, 512, 3), np.uint8)

pts = np.array([[41, 200], [470, 200], [120, 460], [250, 40], [391, 460]])
#print(pts)
pts = pts.reshape(-1, 1, 2)
#print(pts)

cv2.polylines(image, [pts], True, (255, 255, 255), 8)

cv2.imwrite('./images/Drawing-02.png', image)

pt.imshow(image[..., ::-1])
```

```python
import arabic_reshaper

from bidi.algorithm import get_display

from PIL import Image, ImageFont, ImageDraw

image = np.zeros((768,768,3), np.uint8)

base_text = ' '
reshaped_text = arabic_reshaper.reshape(base_text)  #fix separate characters
bidi_text = get_display(reshaped_text) #direction issue (rtl)
display_text = bidi_text.encode().decode('utf-8') #encoding problem (box characters)

pil_image = Image.fromarray(image)

font = ImageFont.truetype('DimaNaskh', 100, encoding='unic')
draw = ImageDraw.Draw(pil_image)
draw.text((50,150), display_text, (0, 200, 150), font=font)

pt.imshow(pil_image)
```

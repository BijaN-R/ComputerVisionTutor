```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

image = cv2.imread('./images/transparency.png', cv2.IMREAD_UNCHANGED)
cvt_image = cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA)
plt.figure(figsize=[15, 5])
plt.subplot(131); plt.imshow(cvt_image[..., :3]); plt.title('Image');
plt.subplot(132); plt.imshow(image[..., 3], 'gray'); plt.title('alpha channel');
plt.subplot(133); plt.imshow(cvt_image); plt.title('Transparent Image');
```

```python
image = cv2.imread('./images/peach.png')

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
mask_1 = hsv_image[..., 2] >= 200 
mask_1 *= hsv_image[..., 1] < 20
transparent_img = cv2.cvtColor(image, cv2.COLOR_BGR2BGRA)
transparent_img[..., 3][mask_1] = 0
mask_2 = np.invert(mask_1)
transparent_img[..., 3][mask_2] = 255
cv2.imwrite('./images/peach-transparent_01.png', transparent_img)
transparent_img = cv2.cvtColor(transparent_img, cv2.COLOR_BGRA2RGBA)
plt.figure(figsize=[10, 5])

plt.subplot(121); plt.imshow(transparent_img); plt.title('result');
plt.subplot(122); plt.imshow(transparent_img[..., 3], 'gray'); plt.title('alpha');
```

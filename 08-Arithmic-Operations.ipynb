```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./images/pastry-dish.jpg')

M = np.ones(image.shape, dtype='uint8') * 100

added = cv2.add(image, M)
subtracted = cv2.subtract(image, M)

plt.figure(figsize=[15, 5])
plt.subplot(131); plt.imshow(added[..., ::-1]); plt.title('Added');
plt.subplot(132); plt.imshow(image[..., ::-1]); plt.title('Normal');
plt.subplot(133); plt.imshow(subtracted[..., ::-1]); plt.title('Subtracted');
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

logo = cv2.imread('./images/Logo.png')
image = cv2.imread('./images/noon_khameii.jpg')

Mlogo = np.ones(logo.shape, dtype='uint8') * 100
logo = cv2.subtract(logo, Mlogo)

blank = np.zeros(image.shape, dtype='uint8')
logo = np.append(logo, blank[logo.shape[0]:, :logo.shape[1]], axis=0)
logo = np.append(blank[:, logo.shape[1]:], logo, axis=1)

image = cv2.add(logo, image) 

plt.imshow(image[..., ::-1])

cv2.imwrite('./images/overlay-logo-01.jpg', image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

logo = cv2.imread('./images/Logo1.png')
image = cv2.imread('./images/noon_khameii.jpg')

logo_h, logo_w, _ = logo.shape

Mlogo = np.ones(logo.shape, dtype='uint8') * 100
logo = cv2.subtract(logo, Mlogo)

blank = np.zeros_like(image)
blank[50:50+logo_h, 700:700+logo_w, :] = logo

image = cv2.add(blank, image) 

plt.imshow(image[..., ::-1])

cv2.imwrite('./images/overlay-logo-02.jpg', image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

logo1 = cv2.imread('./images/Logo1.png')
logo2 = cv2.imread('./images/Logo2.png')

image1 = cv2.addWeighted(logo1, 0.5, logo2, 0.5, 0)
image2 = cv2.addWeighted(logo1, 0.6, logo2, 0.8, -60)

cv2.imwrite('./images/Mix-of-two-logos.png', image2)

plt.figure(figsize=[15, 10])
plt.subplot(231); plt.imshow(logo1[..., ::-1]); plt.title('Logo 1');
plt.subplot(232); plt.imshow(logo2[..., ::-1]); plt.title('Logo 2');
plt.subplot(234); plt.imshow(image1[..., ::-1]); plt.title('Mixed 1');
plt.subplot(235); plt.imshow(image2[..., ::-1]); plt.title('Mixed 2');
```

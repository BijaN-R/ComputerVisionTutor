```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

gamma = 1.6

image_original = cv2.imread('./images/noon_khameii.jpg')
lookup_table = np.empty((1, 256), np.uint8)

for i in range(256):
    #lookup_table[0, i] = np.clip(pow(i/255.0, gamma)* 370.0, 0, 255)
    #lookup_table[0, i] = np.clip(pow(i/255.0, 1/(math.cos(i/15)))* 255.0, 0, 255)
    lookup_table[0, i] = np.clip(24*math.cos((i+60)/40)+i, 0, 255)
    
out = cv2.LUT(image_original, lookup_table)

plt.figure(figsize=[15, 5])
plt.subplot(121); plt.imshow(image_original[..., ::-1]); plt.title('Original');
plt.subplot(122); plt.imshow(out[..., ::-1]); plt.title('Gamma Correction');
```
